```python
from py3d import Vector3
import numpy
def st(log,sim):
    SST=((log-log.M.M).norm()**2).sum()
    SSR=((sim-log.M.M).norm()**2).sum()
    SSr=((log-sim.M.M).norm()**2).sum()
    SSE=((log-sim).norm()**2).sum()
    print(SST,SSR,SSr,SSE)
log=Vector3.Rand(10,30)
log.x=numpy.linspace(0,9,10)
sim=log[0:-3]+Vector3(x=0.2)
dt=sim[0][0][0]-log[0][0][0]
tmp=sim-Vector3(x=dt)
tmp_size=len(tmp)
i=0
while tmp[-1][0][0]<log[-1][0][0]:
    tmp+=Vector3(x=log[i][0][0]-tmp[0][0][0])
    st(log[i:i+tmp_size],tmp)
    i+=1
    print(i)
```

```python
from py3d import Space,Color
s=Space()
log.as_point(Color(r=1)).render(s)
sim.as_point(Color(g=1)).render(s)
```
